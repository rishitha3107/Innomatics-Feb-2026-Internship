```python
# Problem 1: Social Media – Post Engagement Analyzer

def analyze_engagement(likes_list):
    # Calculate total likes using loop
    total_likes = 0
    for like in likes_list:
        total_likes += like

    # Check post status
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    return total_likes, status


# Example Data
likes = [200, 300, 150, 400]
total, post_status = analyze_engagement(likes)

print("Total Likes:", total)
print("Post Status:", post_status)
```

```python
# Problem 3: Agriculture – Rainfall Adequacy Checker

def check_rainfall(rainfall_data, required_level):
    # Calculate average rainfall
    total = 0
    for rain in rainfall_data:
        total += rain

    average = total / len(rainfall_data)

    # Check adequacy
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    return average, status


rainfall = [60, 80, 75, 70]
avg, rainfall_status = check_rainfall(rainfall, 70)

print("Average Rainfall:", avg)
print("Rainfall Status:", rainfall_status)
```

```python
# Problem 4: Social Media – Duplicate Account Detection

def detect_duplicates(usernames):
    # Convert list to set
    unique_users = set(usernames)

    # Check duplicates
    if len(unique_users) < len(usernames):
        return "Yes"
    else:
        return "No"


users = ["user1", "user2", "user1", "user3"]
result = detect_duplicates(users)

print("Duplicate Accounts Found:", result)
```

```python
# Problem 6: Agriculture – Premium Crop Price Filter

def filter_premium_crops(prices):
    premium = []

    # Filter prices above 2000
    for price in prices:
        if price > 2000:
            premium.append(price)

    return premium


crop_prices = [1500, 2500, 1800, 3200]
premium_crops = filter_premium_crops(crop_prices)

print("Premium Crops:", premium_crops)
```

```python
# Problem 7: System Monitoring – Application Health Checker

def check_application_health(error_count):
    # Evaluate error count
    if error_count == 0:
        return "Healthy"
    elif error_count <= 5:
        return "Minor Issues"
    else:
        return "Critical Issues"


errors = 7
status = check_application_health(errors)

print("Error Count:", errors)
print("System Status:", status)
```

```python
# Problem 8: Banking – Daily Transaction Limit Checker

def check_transaction(amount):
    limit = 50000

    # Check transaction limit
    if amount <= limit:
        return "Approved"
    else:
        return "Rejected"


amount = 60000
status = check_transaction(amount)

print("Transaction Amount:", amount)
print("Transaction Status:", status)
```

```python
# Problem 9: E-Learning – Student Attendance Eligibility System

def check_attendance(attendance_list):
    # Calculate attendance percentage
    total_classes = len(attendance_list)
    attended = 0

    for status in attendance_list:
        if status == "Present":
            attended += 1

    percentage = (attended / total_classes) * 100

    # Check eligibility
    if percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"

    return percentage, eligibility


attendance = ["Present", "Absent", "Present", "Present", "Present"]
percent, status = check_attendance(attendance)

print("Attendance Percentage:", percent)
print("Exam Eligibility:", status)
```

```python
# Problem 10: Smart Electricity Bill Analyzer

def calculate_electricity_bill(units):
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    # Usage classification
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"
    return bill, usage
units = 250
total_bill, usage_status = calculate_electricity_bill(units)

print("Total Bill:", total_bill)
print("Usage Status:", usage_status)
```
