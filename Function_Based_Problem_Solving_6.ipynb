```python
# Problem 1: Smart Parking Lot Management System

def parking_system(capacity, logs):
    current = 0

    for log in logs:
        if log == "IN":
            current += 1
        elif log == "OUT" and current > 0:
            current -= 1

    if current > capacity:
        status = "Over Capacity Alert!"
    elif current == capacity:
        status = "Full"
    else:
        status = "Available"

    print("Currently Parked Vehicles:", current)
    print("Parking Status:", status)


# Sample Input
parking_system(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

```python
# Problem 2: Online Food Delivery Time Estimator

def estimate_delivery_time(distance, traffic, weather):
    base_time = distance * 5  # 5 minutes per km

    # Traffic adjustments
    if traffic == "High":
        base_time += 15
    elif traffic == "Medium":
        base_time += 8
    else:
        base_time += 3

    # Weather adjustments
    if weather == "Rainy":
        base_time += 10
    elif weather == "Stormy":
        base_time += 20

    print("Estimated Delivery Time:", base_time, "minutes")

# Sample Input
estimate_delivery_time(8, "High", "Rainy")
```

```python
# Problem 3: Movie Theatre Seat Occupancy Analyzer

def seat_occupancy_analyzer(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100

    # Determine show status
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"

    print("Occupancy:", round(occupancy), "%")
    print("Show Status:", status)


# Sample Input
booked = [1] * 150
seat_occupancy_analyzer(200, booked)
```

```python
 #Problem 4: Cloud Server Load Classification System

def server_load(cpu_readings):
    total = 0
    for reading in cpu_readings:
        total += reading
    avg = total / len(cpu_readings)
    if avg < 50:
        status = "Normal"
    elif avg <= 80:
        status = "Warning"
    else:
        status = "Critical"
    print("Average CPU Load:", round(avg), "%")
    print("Server Status:", status)

server_load([45, 60, 70, 85, 90])
```

```python
# Problem 5: Smart Classroom Resource Usage Monitor

def classroom_resource_monitor(resource_usage):
    overused = []

    # Assume overuse threshold = 8 hours
    for resource, hours in resource_usage.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

classroom_resource_monitor(usage)
```

```python
# Problem 6: Online Event Registration Capacity Controller

def event_registration_controller(capacity, registrations):
    if registrations > capacity:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlisted = 0
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


# Sample Input
event_registration_controller(100, 105)
```
