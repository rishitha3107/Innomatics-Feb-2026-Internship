```python
# Problem 1: Unique Words in a Sentence

sentence = input("Enter sentence: ")

# Splitting sentence into words
words = sentence.split()

# Converting list to set to remove duplicates
unique_words = set(words)

# Printing results
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
# Problem 2: Highest Salary from Employee Data

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Find the employee with the highest salary using the max function and a key
highest_employee = max(employees, key=employees.get)
print("Highest Salary:", highest_employee, "-", employees[highest_employee])
```

```python
# Problem 3: Find Maximum and Minimum Values

numbers = [45, 22, 89, 10, 66]

# Initialize max and min
maximum = numbers[0]
minimum = numbers[0]

for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

```python
# Problem 4: Count Products Above a Price Threshold

prices = [450, 1200, 899, 1500, 300]

count = 0
for price in prices:
    if price > 1000: # Check if the price is above the threshold
        count += 1 # Increment count if condition is met

print("Products above 1000:", count)
```

```python
# Problem 5: Calculate Attendance Percentage

attendance = ["P", "P", "A", "P", "P"]

present_count = 0

for status in attendance:
    if status == "P":
        present_count += 1

percentage = (present_count / len(attendance)) * 100

print("Attendance Percentage:", percentage)
```

```python
# Problem 6: Remove Duplicate Phone Numbers
phone_numbers = [9876543210, 9123456789, 9876543210]

# Convert list to a set to automatically remove duplicates
unique_numbers = set(phone_numbers)
print("Unique phone numbers:", unique_numbers)
```

```python
# Problem 7: Count Character Frequency

text = "python"

frequency = {}

for char in text:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1

print(frequency)
```

```python
# Problem 8: Convert List to Tuple
numbers_list = [10, 20, 30]

# Convert the list to a tuple
numbers_tuple = tuple(numbers_list)
print("Tuple:", numbers_tuple)
```

```python
# Problem 9: Check if a Key Exists in a Dictionary

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

key_to_check = "Ravi"

if key_to_check in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

```python
# Problem 10: Calculate Average Marks
marks = [70, 75, 65, 80]

# Calculate the sum and divide by the number of elements to get the average
average = sum(marks) / len(marks)
print("Average Marks:", average)
```
