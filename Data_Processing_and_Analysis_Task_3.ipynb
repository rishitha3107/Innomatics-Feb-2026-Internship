```python
# Problem Statement 1: Employee Performance Bonus Eligibility

# Given dictionary of employees and their performance scores
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Step 1: Find the highest performance score
highest_score = max(employees.values())

# Step 2: Find all employees who have the highest score
top_performers = []

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

# Step 3: Display result
print("Top Performers Eligible for Bonus:",
      ", ".join(top_performers),
      f"(Score: {highest_score})")
```

```python
# Problem Statement 2: Search Query Keyword Analysis

# Given search query
query = "Buy mobile phone buy phone online"

# Step 1: Convert to lowercase
query = query.lower()

# Step 2: Remove punctuation (if any)
import string
query = query.translate(str.maketrans('', '', string.punctuation))

# Step 3: Split sentence into words
words = query.split()

# Step 4: Count frequency of each word
frequency = {}

for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1

# Step 5: Display only words searched more than once
result = {}

for word, count in frequency.items():
    if count > 1:
        result[word] = count

print(result)
```

```python
# Problem Statement 3: Sensor Data Validation

# Given sensor readings (index = hour, value = reading)
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Step 1: Create an empty list to store valid readings
valid_readings = []

# Step 2: Loop through the list using index and value
for hour, value in enumerate(sensor_readings):

    # Step 3: Check if reading is even (valid)
    if value % 2 == 0:
        valid_readings.append((hour, value))  # Store as tuple (hour, value)

# Step 4: Display result
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
# Problem Statement 4: Email Domain Usage Analysis

# Given list of employee email IDs
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Step 1: Count occurrences of each domain
domain_count = {}

for email in emails:
    domain = email.split("@")[1]   # Extract domain part

    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

# Step 2: Calculate percentage usage
total_emails = len(emails)

print("Email Domain Usage Percentage:")

for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {int(percentage)}%")
```

```python
# Problem Statement 5: Sales Spike Detection

# Given daily sales data
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Step 1: Calculate average daily sales
average_sales = sum(sales) / len(sales)

# Step 2: Define spike threshold (30% above average)
threshold = average_sales * 1.30

# Step 3: Detect days with sales greater than threshold
print("Sales Spike Days:")

for day, value in enumerate(sales, start=1):  # start=1 for day numbering
    if value > threshold:
        print(f"Day {day}: {value}")
```

```python
# Problem Statement 6: Duplicate User ID Detection

# Given list of user IDs
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Step 1: Count frequency of each user ID
frequency = {}

for user in user_ids:
    if user in frequency:
        frequency[user] += 1
    else:
        frequency[user] = 1

# Step 2: Display only duplicates (count > 1)
for user, count in frequency.items():
    if count > 1:
        print(f"{user} → {count} times")
```
